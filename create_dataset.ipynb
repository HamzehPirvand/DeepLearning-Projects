```python
from glob import glob
import imageio
import numpy as np
import matplotlib.pyplot as plt
```

```python
ListOfTrainingImages=glob("./membrane/train/image/*")
ListOfTrainingLabels=glob("./membrane/train/label/*")
```

```python
allTrainingImages=[]
for imagePath in ListOfTrainingImages:
    imageData=np.array(imageio.imread(imagePath))
    allTrainingImages.append(imageData)
    
allTrainingLabel=[]
for imagePath in ListOfTrainingLabels:
    imageData=np.array(imageio.imread(imagePath))
    allTrainingLabel.append(imageData)
```

```python
print(allTrainingImages[0].shape)
plt.imshow(allTrainingImages[0], cmap='gray')
```

```python
allTrainingImages=np.array(allTrainingImages)
```

```python
allTrainingImages.shape
```

```python
print(allTrainingLabel[0].shape)
plt.imshow(allTrainingLabel[0], cmap='gray')
```

```python
allTrainingLabel=np.array(allTrainingLabel)
```

```python
allTrainingLabel.shape
```

```python
import pickle

with open('membraneData.pickle', 'wb') as handle:
    pickle.dump((allTrainingImages, allTrainingLabel), handle, protocol=pickle.HIGHEST_PROTOCOL)
```
